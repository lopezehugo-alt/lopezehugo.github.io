4.1 NumPy (Numerical Python)
NumPy √© fundamental para computa√ß√£o num√©rica em Python. Ele fornece um objeto array multidimensional de alto desempenho e ferramentas para trabalhar com esses arrays.

C√©lula de C√≥digo: Exemplo com NumPy

4.2 Pandas

Pandas √© a biblioteca de fato para manipula√ß√£o e an√°lise de dados. Sua principal estrutura de dados, o DataFrame, √© ideal para trabalhar com dados tabulares (como planilhas do Excel ou tabelas de bancos de dados).

C√©lula de C√≥digo: Exemplo com Pandas

```python
import pandas as pd

# Criando um dicion√°rio com dados
dados = {
    'Nome': ['Jo√£o', 'Maria', 'Pedro', 'Ana'],
    'Idade': [25, 30, 35, 28],
    'Cidade': ['S√£o Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'S√£o Paulo']
}

# Criando um DataFrame a partir do dicion√°rio
df = pd.DataFrame(dados)

print("DataFrame criado:\n", df)

# Filtrando dados: pessoas com mais de 30 anos
df_filtrado = df[df['Idade'] > 30]
print("\nPessoas com mais de 30 anos:\n", df_filtrado)

# Calculando a m√©dia da idade
media_idade = df['Idade'].mean()
print(f"\nM√©dia de idade: {media_idade:.2f}")
```

* Explica√ß√£o do C√≥digo Pandas (Exemplo B√°sico)

Este c√≥digo √© um √≥timo ponto de partida para entender como a biblioteca Pandas funciona, a principal ferramenta para an√°lise de dados em Python. Ele demonstra como criar e manipular uma das suas estruturas de dados mais importantes: o DataFrame.

* Construindo o DataFrame

Primeiro, a biblioteca Pandas √© importada com o apelido pd. Em seguida, um dicion√°rio em Python √© criado para armazenar os dados. As chaves do dicion√°rio (como 'Nome' e 'Idade') se tornam os nomes das colunas, enquanto as listas de valores se tornam as linhas do DataFrame.

A linha mais importante √© onde o DataFrame √© de fato criado a partir desse dicion√°rio. O DataFrame se assemelha a uma tabela ou planilha, com linhas e colunas, tornando a organiza√ß√£o e a manipula√ß√£o de dados muito mais intuitiva.

* Filtragem e Sele√ß√£o de Dados

Uma das capacidades mais √∫teis do Pandas √© a filtragem de dados. No exemplo, uma condi√ß√£o √© usada para selecionar apenas as pessoas com mais de 30 anos. O Pandas avalia essa condi√ß√£o linha por linha, retornando True ou False para cada uma. Em seguida, ele usa essa s√©rie de valores para "filtrar" o DataFrame original, exibindo apenas as linhas onde a condi√ß√£o foi True. Isso permite que voc√™ trabalhe com subconjuntos espec√≠ficos dos seus dados de forma simples e r√°pida.

* Realizando C√°lculos

Por fim, o c√≥digo mostra como √© f√°cil realizar c√°lculos b√°sicos. Ao acessar a coluna 'Idade', podemos usar o m√©todo .mean() para calcular a m√©dia de todos os valores na coluna. # Estudo de Caso

 * An√°lise de Chargeback em E-commerce

Este notebook simula um cen√°rio de an√°lise de dados de vendas, focando em transa√ß√µes que resultaram em chargeback (fraude). Utilizaremos as bibliotecas Pandas para manipula√ß√£o e an√°lise dos dados e Matplotlib para a visualiza√ß√£o dos resultados.

 ‚ñ∂ Prepara√ß√£o do Ambiente e Cria√ß√£o da Base de Dados

Primeiro, vamos importar as bibliotecas necess√°rias e criar nossa base de dados fict√≠cia. Os dados ser√£o gerados de forma aleat√≥ria para simular um cen√°rio real.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
from datetime import datetime

# Definindo listas de dados para a simula√ß√£o
nomes = ['Jo√£o', 'Maria', 'Pedro', 'Ana', 'Carlos', 'Sofia', 'Lucas', 'Laura', 'Gabriel', 'Isabela']
estados = ['SP', 'RJ', 'MG', 'RS', 'SC']
incidentes = ['Chargeback', 'Venda Aceita']
meses = ['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril', 'Maio', 'Junho']
bandeiras = ['Visa', 'Mastercard', 'Elo', 'Amex']
produtos = ['Smartphone X', 'Notebook Pro', 'Fone Bluetooth', 'Smartwatch', 'C√¢mera']

# Criando a base de dados
# Usaremos um dicion√°rio para gerar os dados aleat√≥rios
base_dados = {
    'Nome': [random.choice(nomes) for _ in range(1000)],
    'Idade': [random.randint(18, 60) for _ in range(1000)],
    'UF': [random.choice(estados) for _ in range(1000)],
    'Valor_Transacao': [round(random.uniform(50, 5000), 2) for _ in range(1000)],
    'Incidente': [random.choice(incidentes) for _ in range(1000)],
    'Mes': [random.choice(meses) for _ in range(1000)],
    'Bandeira_Cartao': [random.choice(bandeiras) for _ in range(1000)],
    'Produto': [random.choice(produtos) for _ in range(1000)],
}

# Criando o DataFrame
df = pd.DataFrame(base_dados)

# Criando uma amostra de chargebacks para a an√°lise
# Isso garante que teremos dados para analisar
chargebacks = df.sample(n=100)
chargebacks['Incidente'] = 'Chargeback'
df.update(chargebacks)

print("Amostra do DataFrame criado:")
print(df.head())
print("\nInforma√ß√µes do DataFrame:")
df.info()
```

 ‚ñ∂ An√°lise dos Dados de Chargeback

Agora, vamos filtrar o DataFrame para analisar apenas as transa√ß√µes que tiveram um chargeback. A partir deste novo DataFrame, responderemos √†s perguntas da nossa an√°lise.

 ‚ñ∂ Idade M√©dia dos Fraudadores

```python
# Filtrando o DataFrame para transa√ß√µes com chargeback
df_chargeback = df[df['Incidente'] == 'Chargeback']

# Calculando a idade m√©dia
idade_media_chargeback = df_chargeback['Idade'].mean()

print(f"A idade m√©dia das pessoas envolvidas em chargeback √©: {idade_media_chargeback:.2f} anos")
```

 ‚ñ∂ Estados com Maiores Concentra√ß√µes de Chargeback

```python
# Contando a quantidade de chargebacks por estado
chargebacks_por_uf = df_chargeback['UF'].value_counts()

print("Estados com a maior quantidade de chargebacks:")
print(chargebacks_por_uf)

# Criando o gr√°fico de barras
plt.figure(figsize=(10, 6))
chargebacks_por_uf.plot(kind='bar', color='skyblue')
plt.title('Concentra√ß√£o de Chargebacks por UF')
plt.xlabel('Estado (UF)')
plt.ylabel('N√∫mero de Chargebacks')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

 ‚ñ∂ Produtos Mais Afetados

```python
# Contando a quantidade de chargebacks por produto
chargebacks_por_produto = df_chargeback['Produto'].value_counts()

print("Produtos mais afetados por chargeback:")
print(chargebacks_por_produto)

# Criando o gr√°fico de barras
plt.figure(figsize=(12, 7))
chargebacks_por_produto.plot(kind='bar', color='salmon')
plt.title('Produtos Mais Afetados por Chargeback')
plt.xlabel('Produto')
plt.ylabel('N√∫mero de Chargebacks')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

 ‚ñ∂ An√°lise Temporal: Vendas e Perdas por M√™s

```python
# Agrupando por m√™s para contar o n√∫mero de incidentes
incidente_por_mes = df.groupby(['Mes', 'Incidente']).size().unstack(fill_value=0)

# Reordenando os meses para a ordem correta
ordem_meses = ['Janeiro', 'Fevereiro', 'Mar√ßo', 'Abril', 'Maio', 'Junho']
incidente_por_mes = incidente_por_mes.reindex(ordem_meses)

print("Distribui√ß√£o de incidentes por m√™s:")
print(incidente_por_mes)

# Criando o gr√°fico de barras agrupadas
incidente_por_mes.plot(kind='bar', figsize=(12, 7))
plt.title('An√°lise Temporal de Incidentes por M√™s')
plt.xlabel('M√™s')
plt.ylabel('N√∫mero de Transa√ß√µes')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

 ‚ñ∂ Valor de Perda por M√™s

```python
# Agrupando por m√™s e somando o valor das transa√ß√µes com chargeback
perda_por_mes = df_chargeback.groupby('Mes')['Valor_Transacao'].sum()

# Reordenando os meses
perda_por_mes = perda_por_mes.reindex(ordem_meses, fill_value=0)

print("Valor de perda por m√™s devido a chargeback:")
print(perda_por_mes)

# Criando o gr√°fico de linha
plt.figure(figsize=(12, 7))
perda_por_mes.plot(kind='line', marker='o', color='red', linestyle='--')
plt.title('Valor Total de Perda por M√™s (Chargeback)')
plt.xlabel('M√™s')
plt.ylabel('Valor de Perda (R$)')
plt.grid(axis='y', linestyle='--')
plt.show()
```

 ‚ñ∂ Bandeiras de Cart√£o Mais Afetadas
